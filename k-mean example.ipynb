```python
from sklearn.cluster import KMeans, DBSCAN
from collections import Counter
import cv2
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
def rgb2hex(rgb):
    return '#{:02x}{:02x}{:02x}'.format(int(rgb[0]),int(rgb[1]), int(rgb[2]))
```

```python
def plot_image_info(path, k=6):
    #load img
    img_bgr = cv2.imread(path)
    img_rgb = cv2.cvtColor(img_bgr, cv2.COLOR_BGR2RGB)
    
    #resize for speed up processing time
    img = cv2.resize(img_rgb, (64, 64), interpolation=cv2.INTER_AREA)
    
    #reshape to be a list of pixels
    img_list = img.reshape(img.shape[0] * img.shape[1], 3)
    
    #cluster the pixels & assign label
    clt = KMeans(n_clusters=k)
    labels = clt.fit_predict(img_list)
    
    #count labels to find most popular
    label_count = Counter(labels)
    ##total_count = sum(label_count.values())
    
    #subnet out most popular centroid
    center_colors = list(clt.cluster_centers_)
    ordered_colors = [center_colors[i]/255 for i in label_count.keys()]
    color_labels = [rgb2hex(center_colors[i]) for i in label_count.keys()]
    
    plt.figure(figsize=(15, 15))
    plt.subplot(221)
    plt.imshow(img_rgb)
    plt.axis('off')
    
    plt.subplot(222)
    plt.pie(label_count.values(), labels=color_labels, colors=ordered_colors, startangle=90)
    plt.axis('equal')
    #plt.show()
```
